```python
import pandas as pd
import numpy as np
frailty_raw_data = pd.read_csv('frailty_raw_data.csv')
frailty_raw_data
```

```python
columns = list(frailty_raw_data.columns)
columns
```

```python
columns = list(map(lambda x: x.strip().replace(' ', '_'), columns))
columns
```

```python
frailty_raw_data.columns = columns
```

```python
frailty_raw_data.dtypes
```

```python
frailty_raw_data['Grip_strength'] = frailty_raw_data['Grip_strength'].astype(int)
frailty_raw_data['Frailty'] = frailty_raw_data['Frailty'].astype(str)
```

```python
frailty_raw_data.dtypes
```

```python
frailty_raw_data['Frailty'] = frailty_raw_data['Frailty'].map(lambda x: x.strip())
```

```python
map_v = {'Y': 1, 'N': 0}
frailty_raw_data['Frailty'] = frailty_raw_data['Frailty'].map(map_v)
```

```python
frailty_raw_data.head()
```

```python
frail_group = frailty_raw_data[frailty_raw_data['Frailty'] == 1]
non_frail_group = frailty_raw_data[frailty_raw_data['Frailty'] == 0]

avg_grip_strength_frail = frail_group['Grip_strength'].mean()
avg_grip_strength_non_frail = non_frail_group['Grip_strength'].mean()

print("Average Grip Strength for Frail individuals:", avg_grip_strength_frail)
print("Average Grip Strength for Non-Frail individuals:", avg_grip_strength_non_frail)
```
