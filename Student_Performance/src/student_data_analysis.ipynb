```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data=pd.read_csv('/content/cleaned_data.csv')
```

```python
data
```

```python
figure, axis = plt.subplots(2, 2, figsize=(12, 10))


sns.violinplot(x=data['gender'], y=data['math score'], ax=axis[0, 0], palette='Blues_r')
sns.violinplot(x=data['gender'], y=data['reading score'], ax=axis[0, 1], palette='Reds_r')
sns.violinplot(x=data['gender'], y=data['writing score'], ax=axis[1, 0], palette='Greens_r')
sns.violinplot(x=data['gender'], y=data['Total score'], ax=axis[1, 1], palette='Purples_r')


plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)


plt.suptitle('Scores distribution by gender', fontsize=16)


plt.savefig('/scoresdistribution.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))


plt.bar(data['gender'].value_counts().index,
        data['gender'].value_counts().values,
        color=['yellow', 'red'],
        linewidth=2,
        edgecolor='black')

plt.title('Gender Distribution')
plt.xticks(data['gender'].value_counts().index, ['Female', 'Male'])
plt.savefig('Student_Performance', bbox_inches='tight')
plt.show()
```

```python
male_data = data[data['gender'] == 'male']
female_data = data[data['gender'] == 'female']


plt.scatter(male_data['math score'], male_data['reading score'], color='green', label='Male')
plt.scatter(female_data['math score'], female_data['reading score'], color='black', label='Female')


plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('Math Score Distribution vs Reading Score Distribution')


plt.legend()


plt.savefig('mathvsreadingdistribution.png')


plt.show()
```

```python
figure, axis = plt.subplots(2,2)

sns.histplot(data['math score'],ax=axis[0,0],color='grey',label='math score')
sns.histplot(data['reading score'],ax=axis[0,1],color='green',label='reading score')
sns.histplot(data['writing score'],ax=axis[1,0],color='yellow',label='writing score')
sns.histplot(data['Total score'],ax=axis[1,1],color='red',label='Total score')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Frequency distribution of scores')
plt.savefig('scoresfrequency.png')
```

```python
graph = sns.FacetGrid(data, col ='gender',hue_kws={'color': ['r', 'b']}, hue ='test preparation course')

graph.map(plt.hist, 'Total score', bins = 3, color ='black').add_legend()

plt.savefig('coursecompletiondistribution.png')
```
